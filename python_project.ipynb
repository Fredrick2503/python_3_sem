# Program 4


```python
class array:
    def __init__(self):
        self.data=[]
    def Append(self):
        self.data.append(int(input("Enter the element: ")))
    def Delete(self):
        ele=int(input("Enter the element to be deleted: "))
        if ele in self.data:
            self.data.remove(ele)
        else:
            print("Element not present")
    def Display(self):
        print(self.data)

l1=array()
print("Menu: \n1.Append\n2.Delete\n3.Display\n4.Exit")
while(True):
    choice=int(input("Enter choice(1-4): "))
    if choice==1:
        l1.Append()
    elif choice==2:
        l1.Delete()
    elif choice==3:
        l1.Display()
    elif choice==4:
        break
    else:
        print("Invalid choice")
```

# Program 6


```python
class complexnum():
    def __init__(self,real,imaginary):
        self.real=real
        self.imaginary=imaginary;
    def __add__(self,other):
        return complexnum(self.real+other.real,self.imaginary+other.imaginary)
    def __sub__(self,other):
        return complexnum(self.real-other.real,self.imaginary-other.imaginary)
    def __mul__(self,other):
        real_part=(self.real*other.real)-(self.imaginary*other.imaginary)
        imaginary_part=(self.real*other.imaginary)+(self.imaginary*other.real)
        return complexnum(real_part,imaginary_part)
    def __str__(self):
        return f"{self.real}+{self.imaginary}i"
if __name__=="__main__":
    c1=complexnum(2,3)
    c2=complexnum(1,4)
print(c1+c2)
print(c1-c2)
print(c1*c2)
```

# Program 8

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

data={'Variable_A':[1,2,3,4,5,6],'Variable_B':[2,3,4,5,6,7]}
df=pd.DataFrame(data)

def calculate_correlation(df,col1,col2):
    correlation,_=pearsonr(df[col1],df[col2])
    return correlation

def joint_probability(df,event_A,event_B):
    joint_df=df[(df['Variable_A']==event_A)&(df['Variable_B']==event_B)]
    joint_prob=len(joint_df)/len(df)
    return joint_prob

def marginal_probability(df,event,variable):
    marginal_df=df[df[variable]==event]
    marginal_prob=len(marginal_df)/len(df)
    return marginal_prob

def conditional_probability(df,event_A,event_B):
    joint_prob=joint_probability(df,event_A,event_B)
    marginal_prob_B=marginal_probability(df,event_B,'Variable_B')
    if marginal_prob_B==0:
        return 0
    return joint_prob/marginal_prob_B

correlation=calculate_correlation(df,'Variable_A','Variable_B')
print("Correlation between Variable_A,Variable_B: ",correlation)

joint_prob=joint_probability(df,3,4)
print("Joint Probability P(A=3 and B=4): ",joint_prob)

marginal_prob_A=marginal_probability(df,3,'Variable_A')
print("Marginal Probabilty P(A=3): ",marginal_prob_A)

marginal_prob_B=marginal_probability(df,3,'Variable_B')
print("Marginal Probabilty P(B=3): ",marginal_prob_B)

cond_prob=conditional_probability(df,3,4)
print("Conditional Probability P(A=1|B=4): ",cond_prob)
```

# Program 9

```python
import pandas as pd

data={'Name':['Alice','Bob','Charlie',None,'Eve'],'Age':[25,None,30,22,None],'City':['New York','Los Angeles','Chicago','Miami',None]}
df=pd.DataFrame(data)

def analyze_missing_data(df):
    print("DataFrame:\n",df)
    print("\nMissing data analysis: ")
    missing_data=df.isnull().sum()
    print("Missing values in each column: ",missing_data[missing_data>0])
    missing_percentage=(missing_data/len(df))*100
    print("Percentage of missing values in each column: ",missing_percentage[missing_percentage>0])
    return missing_data,missing_percentage
def handle_missing_data(df):
    print("\nChoose a method to handle missing data: ")
    print("1.Filling missing values with specified value")
    print("2.Drop rows with missing values")
    print("3.Drop columns with missing values")
    choice=input("Enter choice(1-3): ")
    if choice=='1':
        value=input("Enter value to fill missing data: ")
        df_filled=df.fillna(value)
        print("\nData after filling missing values: \n",df_filled)
    elif choice=='2':
        df_dropped_rows=df.dropna()
        print("\nData after dropping rows with missing value: ",df_dropped_rows)
    elif choice=='3':
        df_dropped_columns=df.dropna(axis=1)
        print("\nData after dropping columns with missing value: ",df_dropped_columns)
    else:
        print("Invalid choice")
missing_data,missing_percentage=analyze_missing_data(df),handle_missing_data(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    'Age':[28,31,35,40,45,50,55,69,65,70],
    'Salary':[20000,25000,30000,35000,40000,45000,50000,55000,60000,65000],
    'Department':['HR','IT','Finance','IT','Finance','HR','HR','Finance','IT','Finance']
}
df=pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.scatter(df['Age'],df['Salary'],color='blue',marker='o')
plt.title("Sactter plot of Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df['Age'],bins=5,color='green',edgecolor='black')
plt.title("Histogram of age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(8,5))
department_count=df['Department'].value_counts()
plt.bar(department_count.index,department_count.values,color='orange',edgecolor='black')
plt.title("Bar plot of Department counts")
plt.xlabel("Department")
plt.ylabel("Numbetr of Employees")
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data={
    'Age':[28,31,35,40,45,50,55,69,65,70],
    'Salary':[20000,25000,30000,35000,40000,45000,50000,55000,60000,65000],
    'Department':['HR','IT','Finance','IT','Finance','HR','HR','Finance','IT','Finance']
}
df=pd.DataFrame(data)
sns.set(style="whitegrid")

plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Age',y='Salary',palette='viridis',hue='Age',s=100)
plt.title("Sactter plot of Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(title='Department')
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(data=df,x='Age',bins=5,kde=True,color="purple")
plt.title("Histogram of Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
sns.countplot(data=df,x='Department',palette='pastel',hue='Department')
plt.title("Bar plot of Department counts")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()
```
